```python
import pyseldon as pd
```

```python
opset = pd.OutputSettings(n_output_network = 20,n_output_agents = 1, print_progress = True, output_initial = True, start_output = 1, start_numbering_from = 0)
```

```python
degroot_settings = pd.DeGrootSettings(20, 1)
```

```python
net_settings = pd.InitialNetworkSettings(number_of_agents = 300, connections_per_agent = 10)
```

```python
sim_options = pd.SimulationOptions(model_string = "DeGroot", model_settings = degroot_settings, output_settings = opset, network_settings = net_settings)
```

```python
simulation = pd.Simulation(options = sim_options)
simulation.run()
```

```python
network = simulation.network
opinions = []
for i in range(network.n_agents()):
  opinions.append(network.agent[i].data.opinion)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(opinions, 'o', label='Agent Opinions')
plt.xlabel('Agent Index')
plt.ylabel('Opinion')
plt.title('Opinions of Agents in the Network')
plt.legend()
plt.grid(True)
plt.show()
```
